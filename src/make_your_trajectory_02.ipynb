```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creat a sequence of trajectories
def constant_velocity_trajectory(start, angle, velocity, dt, duration):
    """
    Generate a trajectory with constant velocity.
    :param start: Starting point of the trajectory (x, y).
    :param rad_direction: Direction of the trajectory in radians.
    :param velocity: Velocity of the trajectory.
    :param dt: Time step for the trajectory.
    :param duration: Duration of the trajectory.
    :return: Array of points representing the trajectory.
    """
    num_points = int(duration / dt)
    x = np.zeros(num_points)
    y = np.zeros(num_points)
    
    for i in range(num_points):
        x[i] = start[0] + i * dt * velocity * np.cos(np.radians(angle))
        y[i] = start[1] + i * dt * velocity * np.sin(np.radians(angle))

    return x, y

# plot the trajectories
start = (0, 0)
angle = 45
velocity = 1
dt = 0.1
duration = 10
x, y = constant_velocity_trajectory(start, angle, velocity, dt, duration)
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Trajectory', color='blue')
plt.title('Constant Velocity Trajectory')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.xlim(-1, 11)
plt.ylim(-1, 11)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────────────────────────────────────
# USER-EDITABLE PARAMETERS
#─────────────────────────────────────────────────────────────────────────────
g0        = 1.0      # base speed [m/s]
yaw_rate2 = 1.0      # rad/s for segment 2 (CCW spin while moving)
yaw_rate4 = -np.pi/4 # rad/s for segment 4 (negative for CW)
A9        = 1.5      # amplitude for oscillating heading in segment 9
cycles9   = 3        # number of sine oscillations in segment 9

# Waypoints for each segment:
p1 = (-2,  0)
p2 = (-2,  1)
p3 = (3,  1)
p4 = (4,  1)
p5 = (4, -1)
p6 = (1.5,-1)
p7 = (-2, -1)
p8 = (-2,  -0.5)
p9 = (4,  0)
#──────────────────────────────────────────────────────────────────────────────

# Compute segment durations (constant‐velocity segments)
d1 = abs(p2[1] - p1[1]) / g0
d2 = np.hypot(p3[0] - p2[0], p3[1] - p2[1]) / g0
d3 = abs(p4[0] - p3[0]) / g0
d4 = abs(np.pi/2) / abs(yaw_rate4)
d5 = abs(p5[1] - p4[1]) / g0 -1
dist6 = abs(p6[0] - p5[0])
d6 = dist6 / ((1 + 2) / 2)+1      # average v=(1+2)/2
d7 = abs(p7[0] - p6[0]) / ((2 + 1) / 2)
d8 = abs(p8[1] - p7[1]) / g0
d9 = abs(p9[0] - p8[0]) / g0*2.25

# Build full time vector
dt    = 0.01
fs    = 1.0 / dt
T_all = d1 + d2 + d3 + d4 + d5 + d6 + d7 + d8 + d9
tsim  = np.arange(0, T_all, dt)

# Preallocate arrays
v_x = np.zeros_like(tsim)
v_y = np.zeros_like(tsim)
psi = np.zeros_like(tsim)

# Helper: time → index
def idx(t): return int(np.round(t * fs))

# Track current index
cur = 0

# ── Segment 1: (0,0)->(0,2), positive Y body, heading=0 ───────────────────
end = cur + idx(d1)
v_x[cur:end] = 0.0
v_y[cur:end] = +g0
psi[cur:end] = 0.0
cur = end

# ── Segment 2: (0,2)->(4,2), global g0, yaw_rate2, local v_body changes ───
end = cur + idx(d2)
t  = np.arange(0, d2, dt)
# global velocity = [g0, 0], body = R(-psi)*v_global
psi[cur:end] = yaw_rate2 * t
v_x[cur:end] =  g0 * np.cos(psi[cur:end])
v_y[cur:end] = -g0 * np.sin(psi[cur:end])
cur = end

# ── Segment 3: (4,2)->(5,2), heading=0, vx=g0 ────────────────────────────
end = cur + idx(d3)
v_x[cur:end] = +g0
v_y[cur:end] = 0.0
psi[cur:end] = 0.0
cur = end

# ── Segment 4: smooth turn 0→π/2, v_body_x=g0, v_y=0 ───────────────────────
end = cur + idx(d4)
t   = np.arange(0, d4, dt)
psi[cur:end] = 0 + yaw_rate4 * t
v_x[cur:end] = +g0
v_y[cur:end] = 0.0
cur = end

# ── Segment 5: move to (5,-2), heading=π, body+Y direction ─────────────
end = cur + idx(d5)
v_x[cur:end] = 0.0
v_y[cur:end] = +g0
psi[cur:end] = np.pi
cur = end

# ── Segment 6: (5,-2)->(2.5,-2), heading=π, accel 1→2 ───────────────────
end = cur + idx(d6)
t   = np.linspace(0, 1, end-cur)  # parameter [0,1]
v_body_x = 1.0 + (2.0-1.0)*t       # ramp 1→2
v_x[cur:end] = v_body_x
v_y[cur:end] = 0.0
psi[cur:end] = np.pi
cur = end

# ── Segment 7: (2.5,-2)->(0,-2), heading=π, decel 2→1 ───────────────────
end = cur + idx(d7)
t   = np.linspace(0, 1, end-cur)
v_body_x = 2.0 + (1.0-2.0)*t       # ramp 2→1
v_x[cur:end] = v_body_x
v_y[cur:end] = 0.0
psi[cur:end] = np.pi
cur = end

# ── Segment 8: (0,-2)->(0,0), heading=π/2, body-forward along global+y ──
end = cur + idx(d8)
v_x[cur:end] = +g0
v_y[cur:end] = 0.0
psi[cur:end] = np.pi/2
cur = end

# ── Segment 9: (0,0)->(5,0), heading oscillates A*sin(w t) ──────────────
end = cur + idx(d9)
t   = np.arange(0, d9, dt)
w   = 2*np.pi*cycles9 / d9/1.5
psi[cur:end] = A9 * np.sin(w * t)
v_x[cur:end] = +g0
v_y[cur:end] = 0.0
cur = end

# ── get z coordinate ────────────────────────────────────────────────
z = 2 * np.ones_like(v_x)

# ── Plotting ─────────────────────────────────────────────────────────────
plt.figure(figsize=(8, 3))
plt.plot(tsim, v_x, label='v_x (body)')
plt.title('Body‐Frame Velocity')
plt.xlabel('Time [s]'); plt.ylabel('Velocity [m/s]')
plt.legend(); plt.grid(); plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 3))
plt.plot(tsim, v_y, label='v_y (body)')
plt.title('Body‐Frame Velocity')
plt.xlabel('Time [s]'); plt.ylabel('Velocity [m/s]')
plt.legend(); plt.grid(); plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 3))
plt.plot(tsim, psi, label='ψ (yaw)')
plt.title('Heading ψ')
plt.xlabel('Time [s]'); plt.ylabel('Yaw [rad]')
plt.grid(); plt.tight_layout()

plt.show()
```

```python
# # convert from body to global coordinates
v_x_global = v_x * np.cos(psi) - v_y * np.sin(psi)
v_y_global = v_x * np.sin(psi) + v_y * np.cos(psi)
# wrap psi between -2pi and 2pi
psi = np.arctan2(np.sin(psi), np.cos(psi))

yaw_rate = np.gradient(psi, dt)

# plot the global x velocity
plt.figure(figsize=(10, 5))
plt.plot(tsim, v_x_global, label='v_x_global', color='blue')
plt.title('Global Velocity in X Direction')
plt.xlabel('Time [s]')
plt.ylabel('Velocity [m/s]')
plt.xlim(0, tsim[-1])
plt.ylim(-2, 2)
plt.show()
# plot the global y velocity
plt.figure(figsize=(10, 5))
plt.plot(tsim, v_y_global, label='v_y_global', color='blue')
plt.title('Global Velocity in Y Direction')
plt.xlabel('Time [s]')
plt.ylabel('Velocity [m/s]')
plt.xlim(0, tsim[-1])
plt.ylim(-2, 2)
plt.show()

#plot heading
plt.figure(figsize=(10, 5))
plt.plot(tsim, psi, label='Heading', color='blue')
plt.title('Heading')
plt.xlabel('Time [s]')
plt.ylabel('Heading [rad]')
plt.xlim(0, tsim[-1])
plt.show()
```

```python
from scipy.integrate import cumtrapz

# Initial position
x0, y0 = -3, 0

# Integrate velocity to get position
x_position = x0 + cumtrapz(v_x_global, tsim, initial=0)
print(x_position)
y_position = y0 + cumtrapz(v_y_global, tsim, initial=0)

# plot x position
plt.figure(figsize=(10, 5))
plt.plot(tsim, x_position, label='x_position', color='blue')
plt.title('X Position')
plt.xlabel('Time [s]')
plt.ylabel('X Position [m]')
plt.xlim(0, tsim[-1])
plt.show()
# plot y position
plt.figure(figsize=(10, 5))
plt.plot(tsim, y_position, label='y_position', color='blue')
plt.title('Y Position')
plt.xlabel('Time [s]')
plt.ylabel('Y Position [m]')
plt.xlim(0, tsim[-1])
plt.show()


# Plot the position trajectory
plt.figure(figsize=(10, 5))
plt.plot(x_position, y_position, label='Position Trajectory', color='black')
plt.title('Position Trajectory from Velocity')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid()
# add box representing the walls at x = 5 x = -5  y = 3 y = -3
plt.axvline(x=5, color='red', linestyle='--', label='Wall at x=5')
plt.axvline(x=-5, color='red', linestyle='--', label='Wall at x=-5')
plt.axhline(y=3, color='red', linestyle='--', label='Wall at y=3')
plt.axhline(y=-3, color='red', linestyle='--', label='Wall at y=-3')

# add a triangle every 10 time steps with the tip at the current position with heading angle
for i in range(0, len(tsim), 25):
    x = x_position[i]
    y = y_position[i]
    angle = psi[i]
    triangle = np.array([[.1, 0], [-0.05, -0.05], [-0.05, 0.05]])*2
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
    rotated_triangle = triangle @ rotation_matrix.T
    # make each segment of the triangle a different color
    if i <d1*fs:
        color = 'blue'
    elif i < (d1 + d2) * fs:
        color = 'green'
    elif i < (d1 + d2 + d3) * fs:
        color = 'red'
    elif i < (d1 + d2 + d3 + d4) * fs:
        color = 'purple'
    elif i < (d1 + d2 + d3 + d4 + d5) * fs:
        color = 'orange'
    elif i < (d1 + d2 + d3 + d4 + d5 + d6) * fs:
        color = 'brown'
    elif i < (d1 + d2 + d3 + d4 + d5 + d6 + d7) * fs:
        color = 'brown'
    elif i < (d1 + d2 + d3 + d4 + d5 + d6 + d7 + d8) * fs:
        color = 'gray'
    else:
        color = 'cyan'
    plt.fill(x + rotated_triangle[:, 0], y + rotated_triangle[:, 1], color=color, alpha=0.5)
    # plt.fill(x + rotated_triangle[:, 0], y + rotated_triangle[:, 1], color='blue', alpha=0.5)
# plt.xlim(2.5, 7.5)
# plt.ylim(2.5, 7.5)
# add black rectangle at x = 5.5 to x = 6 and y = -3 to y = 3
plt.fill_betweenx([-3, 3], 5.5, 6, color='black', alpha=0.5)
plt.xlim(-5.5, 7.0)
plt.ylim(-3.5, 3.5)
# add a column of black arrows coming from the black box to the left with magnitude .75 spaced every .25
for y in np.arange(-2.75, 3, 0.25):
    plt.arrow(5.5, y, -0.5, 0, head_width=0.1, head_length=0.1, fc='black', ec='black')
plt.show()
```
